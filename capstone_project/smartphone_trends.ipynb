Create New Columns:
Camera Total MP: Front Camera + Back Camera (to measure total camera power).
Price-to-Spec Ratio: (RAM + Battery Capacity) / Price (higher = better value).


```python
# Function to extract and sum numeric values from a string
def extract_and_sum_numbers(text):
    numbers = pd.Series(text).str.findall(r'(\d+)').apply(lambda x: sum(map(int, x)) if x else 0)
    return numbers

# Apply the function to the camera columns
phones['Front Camera'] = extract_and_sum_numbers(phones['Front Camera'])
phones['Back Camera'] = extract_and_sum_numbers(phones['Back Camera'])

# Calculate Camera Total MP
phones['Camera Total MP'] = phones['Front Camera'] + phones['Back Camera']

# Append "MP" to the Camera Total MP values
phones['Camera Total MP'] = phones['Camera Total MP'].astype(str) + " MP"
phones.head()
```

```python
# Exchange rates
exchange_rates = {
    'Pakistan': 280,  # 1 USD = 280 PKR
    'India': 83,      # 1 USD = 83 INR
    'China': 7.2,     # 1 USD = 7.2 CNY
    'Dubai': 3.67,    # 1 USD = 3.67 AED
    'USA': 1.0        # Already in USD
}

# Function to clean and convert prices
def clean_and_convert_price(price_series, exchange_rate):
    # Extract numeric values from the price column
    price_series = price_series.astype(str).str.replace(r'[^\d.]', '', regex=True)
    
    # Convert to numeric
    price_series = pd.to_numeric(price_series, errors='coerce')
    # Replace NaN values with mean of that country's price column
    mean_price = price_series.mean()
    price_series.fillna(mean_price, inplace=True)

    # Convert to USD
    price_series = (price_series / exchange_rate).round(2)
    
    return price_series

# Apply function to each price column
for country, rate in exchange_rates.items():
    col_name = f"Launched Price ({country})"
    if col_name in phones.columns:
        phones[f"Price in USD ({country})"] = clean_and_convert_price(phones[col_name], rate)

# Display the updated DataFrame
phones.head()
```

```python
# Function to determine price tiers based on 25th, 50th, and 75th percentiles
def price_tier(price, price_25, price_50, price_75):
    if price < price_25:
        return 'Budget'
    elif price < price_50:
        return 'Mid-Range'
    elif price < price_75:
        return 'Upper Mid-Range'
    else:
        return 'Premium'

# Loop through each country's price column
for country in exchange_rates.keys():
    col_name = f"Price in USD ({country})"
    
    if col_name in phones.columns:
        # Calculate percentiles for this country's price
        price_25 = phones[col_name].quantile(0.25)
        price_50 = phones[col_name].median()  # 50th percentile (median)
        price_75 = phones[col_name].quantile(0.75)
        
        # Apply tier classification
        phones[f"Price Tier ({country})"] = phones[col_name].apply(lambda x: price_tier(x, price_25, price_50, price_75))

# Display the result
phones.head()
```

1. Price Variation Across Countries: "How Geography Shapes Smartphone Prices"
Compare how prices differ across Pakistan, India, China, Dubai, and the USA.
Identify the most expensive and cheapest country for smartphones.
Check if certain brands are consistently priced higher or lower in specific regions.
Investigate currency-adjusted price trends to see if one country gets better deals than others.

```python
# List of countries to compare
countries = ['Pakistan', 'India', 'China', 'Dubai', 'USA']

# Create a DataFrame for the prices in USD for each country
price_data = phones[[f"Price in USD ({country})" for country in countries]]
price_data
```

```python
average_prices=price_data.mean(axis=0)
```

```python
# Sort the countries by average price to find the best overall deals
sorted_average_prices = average_prices.sort_values()

# Print the results
print("Countries ranked by average smartphone price (from cheapest to most expensive):")
for country, avg_price in sorted_average_prices.items():
    # Extract the country name from the column name
    country_name = country.replace('Price in USD (', '').replace(')', '')
    print(f"{country_name}: Average price ${avg_price:.2f}")

# Print summary statement
cheapest_country = sorted_average_prices.index[0].replace('Price in USD (', '').replace(')', '')
most_expensive_country = sorted_average_prices.index[-1].replace('Price in USD (', '').replace(')', '')
print(f"\n{cheapest_country} offers the cheapest smartphones on average, while {most_expensive_country} is the most expensive market.")
```

```python
# Plot the bar graph
plt.figure(figsize=(12, 6))
average_prices.plot(kind='bar')
plt.title('Average Price Comparison Across Countries')
plt.ylabel('Average Price in USD')
plt.xlabel('Country')
plt.xticks(ticks=range(len(average_prices.index)), labels=average_prices.index, rotation=45)
plt.show()
```

```python
# Add the 'Brand' column to the price_data DataFrame
price_data['Model Name'] = phones['Model Name']
price_data['Company Name']= phones['Company Name']
price_data
```

```python
# Melt the DataFrame to have 'Country' as a variable
melted_price_data = price_data.melt(id_vars='Model Name', var_name='Country', value_name='Price')

# Remove the 'Price in USD (' and ')' from the 'Country' column
melted_price_data['Country'] = melted_price_data['Country'].str.replace('Price in USD (', '').str.replace(')', '')

# Group by country and brand, then calculate the average price
average_prices_by_Model_Name = melted_price_data.groupby(['Country', 'Model Name'])['Price'].mean().unstack()
average_prices_by_Model_Name
```

```python
# For each phone model, find the most expensive and cheapest country and the percentage difference between them
for model_name in average_prices_by_Model_Name.columns:
    model_prices = average_prices_by_Model_Name[model_name].dropna()
    most_expensive_country = model_prices.idxmax()
    cheapest_country = model_prices.idxmin()
    max_price = model_prices.max()
    min_price = model_prices.min()
    percentage_difference = ((max_price - min_price) / min_price) * 100
    
    print(f"\nModel: {model_name}")
    print(f"Most expensive country: {most_expensive_country} with price ${max_price:.2f}")
    print(f"Cheapest country: {cheapest_country} with price ${min_price:.2f}")
    print(f"Percentage difference: {percentage_difference:.2f}%")
```

```python
# Melt the DataFrame to have 'Country' as a variable
melted_price_data_by_company = price_data.melt(id_vars=['Model Name', 'Company Name'], var_name='Country', value_name='Price')

# Remove the 'Price in USD (' and ')' from the 'Country' column
melted_price_data_by_company['Country'] = melted_price_data_by_company['Country'].str.replace('Price in USD (', '').str.replace(')', '')

# Group by country and company, then calculate the average price
average_prices_by_company = melted_price_data_by_company.groupby(['Country', 'Company Name'])['Price'].mean().unstack()

# Print the average prices by company for each country
for country in countries:
    print(f"\nAverage prices in {country}:")
    country_prices = average_prices_by_company.loc[country]
    for company, price in country_prices.items():
        print(f"{company}: ${price:.2f}")
```

```python
# For each company, find the most expensive and cheapest country and the percentage difference between them
for company_name in average_prices_by_company.columns:
    company_prices = average_prices_by_company[company_name].dropna()
    most_expensive_country = company_prices.idxmax()
    cheapest_country = company_prices.idxmin()
    max_price = company_prices.max()
    min_price = company_prices.min()
    percentage_difference = ((max_price - min_price) / min_price) * 100
    
    print(f"\nCompany: {company_name}")
    print(f"Most expensive country: {most_expensive_country} with price ${max_price:.2f}")
    print(f"Cheapest country: {cheapest_country} with price ${min_price:.2f}")
    print(f"Percentage difference: {percentage_difference:.2f}%")
```

```python
# For each company, calculate the price range (max - min) across countries
price_ranges_by_company = {}

for company_name in average_prices_by_company.columns:
    company_prices = average_prices_by_company[company_name].dropna()
    max_price = company_prices.max()
    min_price = company_prices.min()
    price_range = max_price - min_price
    price_ranges_by_company[company_name] = price_range

# Sort the companies by price range to find the most and least consistent pricing
sorted_price_ranges = sorted(price_ranges_by_company.items(), key=lambda x: x[1])

# Print the results
print("Brands with the most consistent pricing across regions:")
for company, price_range in sorted_price_ranges[:10]:  # Top 5 most consistent
    print(f"{company}: Price range ${price_range:.2f}")

print("\nBrands with the least consistent pricing across regions:")
for company, price_range in sorted_price_ranges[-1:-11:-1]:  # Top 5 least consistent
    print(f"{company}: Price range ${price_range:.2f}")
```

```python
# For each model, calculate the price range (max - min) across countries
price_ranges_by_model = {}

for model_name in average_prices_by_Model_Name.columns:
    model_prices = average_prices_by_Model_Name[model_name].dropna()
    max_price = model_prices.max()
    min_price = model_prices.min()
    price_range = max_price - min_price
    price_ranges_by_model[model_name] = price_range

# Sort the models by price range to find the most and least consistent pricing
sorted_price_ranges_models = sorted(price_ranges_by_model.items(), key=lambda x: x[1])

# Print the results
print("Models with the most consistent pricing across regions:")
for model, price_range in sorted_price_ranges_models[:10]:  # Top 10 most consistent
    print(f"{model}: Price range ${price_range:.2f}")

print("\nModels with the least consistent pricing across regions:")
for model, price_range in sorted_price_ranges_models[-1:-11:-1]:  # Top 10 least consistent
    print(f"{model}: Price range ${price_range:.2f}")
```
